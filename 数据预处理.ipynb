```python
#创建一个人工数据集, 存在csv文件里, csv文件的意思是comma-separated values,用逗号分开的格式的文本

import os #operating system,提供了和操作系统交互的功能

os.makedirs(os.path.join('..', 'data'),exist_ok=True) #用来创建文件夹
#os.path.join 函数来构建一个文件路径。.. 表示当前目录的上一级目录
#exist_ok=True：这个参数告诉 os.makedir 函数，如果文件夹已经存在，不要抛出异常。如果 exist_ok 设置为 False（默认值），
#并且文件夹已经存在，那么 os.makedir 函数会抛出一个 FileExistsError 异常。

data_file = os.path.join('..', 'data', 'house_tiny.csv') #用逗号链接不同的函数的参数,这是标准形式, ".."表示链接上一层目录,每个参数代表一个路径
with open(data_file, 'w') as f: #用with语句 打开一个文件, w表示write 写入,如果没有文件, 就创建新的再写入
  f.write('NumRooms,Alley,Price\n') #列名
  f.write('NA,Pave,127500\n') #每行一个数据
  f.write('2,NA,106000\n')
  f.write('4,NA,178100\n')
  f.write('NA,NA,140000\n')
```

```python
#读取csv文件用pandas库, 没有装的话就是pip安装
import pandas as pd
data = pd.read_csv(data_file) #函数read_csv
print(data)
#如果不print, 会显示html的形式更紧凑
data
```

```python
#补缺失的数据,如果有缺失的数据,就丢掉or插值, 我们会选择插值
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2] #data.iloc用于分割和提取数据,interger-location based index,基于整数位置的索引
#把NAN的值用剩下数值的均值代替,它使用了 Pandas 库的 fillna() 方法，将 inputs 数据中的缺失值（NaN）替换为对应列的平均值。
#如何判断函数还是方法, fillna就是对象.方法名()的一种方法
inputs = inputs.fillna(inputs.mean(numeric_only=True)) #这里和教材不一样,当 numeric_only=True 时，函数或方法会忽略非数值型数据（例如字符串、日期等），只对数值型数据进行计算或操作
print(inputs)
```

```python
#对于非数值的怎么处理呢, 有一个办法,把缺失的作为数值成为一个特别的类, 把不同种类的数值变成一个特征
inputs = pd.get_dummies(inputs, dummy_na=True)#, dtype=int) #用dummy_na把缺失值单独分类出来了 #和教材不一致, 要多加一个dtype设置为int然后重新执行上一个代码块
print(inputs)
```

```python
#所有的元素都是数值类型了, 可以转成tensor了,转成torch中的tensor才能做机器学习
import torch

X, Y = torch.tensor(inputs.to_numpy(dtype=float)), torch.tensor(outputs.to_numpy(dtype=float))
X, Y
```
